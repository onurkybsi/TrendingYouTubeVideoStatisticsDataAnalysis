```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sn
```

```python
dataWarehouse = pd.read_csv("arrangedDataset.csv")[["video_id","trending_date","title","channel_title","category","category_id","publish_time","tags","views","likes","dislikes","comment_count","thumbnail_link","comments_disabled","ratings_disabled","video_error_or_removed","description","country_abb"]]
```

```python
# Creating required data

dislikesGreaterThanlikes = pd.DataFrame(dataWarehouse, columns=["likes","dislikes","comments_disabled","ratings_disabled"])

dislikesGreaterThanlikes = dislikesGreaterThanlikes[dislikesGreaterThanlikes["dislikes"] > dislikesGreaterThanlikes["likes"]]
print(len(dislikesGreaterThanlikes))

commentsDisabled = dislikesGreaterThanlikes[dislikesGreaterThanlikes["comments_disabled"] == True]

print(len(dislikesGreaterThanlikes)-len(commentsDisabled))
print(len(commentsDisabled))
```

```python
plt.pie([len(dislikesGreaterThanlikes)-len(commentsDisabled), len(commentsDisabled)], labels=["Comments enabled","Comments disabled"], explode=[0,0.2], shadow=True, autopct="%1.1f%%")
plt.title("Relation between dislikes and comments disabled state")
plt.show()
```
